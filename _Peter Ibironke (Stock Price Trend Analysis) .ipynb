```python
#Task 1 — Create the Variables

#Daily Stock Market Data

company_name = "NovaTech Plc"
stock_symbol = "NVT"
opening_price = 152.50
current_price = 168.75
shares_owned = 350
target_price = 170
expected_growth = 8
actual_growth= 10
market_open = True
```

```python
#Task 2 — Calculate the Price Increase

#Increase in stock price
price_increase = current_price - opening_price

print("Increase in Stock Price:", price_increase)
```

```python
#Task 3 — Calculate the Current Investment Value

#Total Value of Investors Shares
investment_value = current_price * shares_owned

print("Total Investment Value:", investment_value)
```

```python
#Task 4 — Update the Current Market Price

#Increase in stock price by 3.25
current_price += 3.25

print("Updated Stock Price:", current_price)
```

```python
#Task 5 — Check if the Target Price Has Been Reached

#has the target price of 170 been achieved
achieved_target = current_price >= target_price

print("Target Price Reached:", achieved_target)
```

```python
#Task 6 — Check if the Stock Performed Above Expectations

# stock Performance check
stock_performance = actual_growth >= expected_growth

print("Stock Performed Above Expectations:", stock_performance)
```

```python
#Task 7 — Determine if the Investment Looks Strong

# Check Strength of Investment
investment_strength = market_open and current_price >= target_price

print("Investment is Strong:", investment_strength)
```

```python
#Task 8 — Determine if Investor Attention Is Required

# Investor attention Required?
investor_attention = not investment_strength or not stock_performance 

print("Investor Attention Required:", investor_attention)
```

```python
#Task 9 — Generate the Stock Performance Report

print()
print("=" * 70)
print("                  DAILY STOCK PERFORMANCE REPORT")
print("=" * 70)
print()
print("Company Name: ", company_name)
print("Stock Symbol: ", stock_symbol)
print("Opening Price: ₦", opening_price)
print("Updated Stock Price: ₦", current_price)
print("Increase in Stock Price: ₦", price_increase)
print("No of Shares:", shares_owned)
print("Total Investment Value: ₦", investment_value)
print("Investor Target Price: ₦", target_price)
print("Expected Growth: ", expected_growth,"%")
print("Actual Growth: ", actual_growth,"%")
print("Target Price Achieved:", achieved_target)
print("Performed Above Expectations:", stock_performance)
print("Market Open:", market_open)
print("Investment is Strong:", investment_strength)
print("Investor Attention Required:", investor_attention)
print("=" * 70)
print()
```
